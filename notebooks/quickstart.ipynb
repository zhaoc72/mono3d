# Mono3D Quick Start

This notebook demonstrates basic usage of the Mono3D framework.

```python
import sys
sys.path.insert(0, '../src')

import torch
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path

from mono3d import (
    GaussianModel,
    visualize_pointcloud,
    save_pointcloud,
)
```

## 1. Create a Gaussian Model

```python
# Initialize model
model = GaussianModel(
    num_gaussians=1000,
    sh_degree=3
)

# Create random point cloud
points = torch.randn(1000, 3)
colors = torch.rand(1000, 3)

# Initialize from point cloud
model.initialize_from_pointcloud(points, colors=colors)

print(f"Initialized {model.num_gaussians} gaussians")
```

## 2. Render the Model

```python
# Render
rendered = model.render()

# Display
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

axes[0].imshow(rendered['color'].permute(1, 2, 0).cpu().numpy())
axes[0].set_title('Color')
axes[0].axis('off')

axes[1].imshow(rendered['depth'][0].cpu().numpy(), cmap='turbo')
axes[1].set_title('Depth')
axes[1].axis('off')

axes[2].imshow(rendered['alpha'][0].cpu().numpy(), cmap='gray')
axes[2].set_title('Alpha')
axes[2].axis('off')

plt.tight_layout()
plt.show()
```

## 3. Convert to Point Cloud

```python
# Convert to point cloud
pc = model.to_pointcloud()

print(f"Point cloud shape: {pc['points'].shape}")
print(f"Colors shape: {pc['colors'].shape}")
```

## 4. Visualize Point Cloud

```python
# Visualize (this will create a matplotlib figure)
visualize_pointcloud(
    pc['points'],
    colors=pc['colors'],
    point_size=10,
    title="Gaussian Model as Point Cloud"
)
```

## 5. Save Results

```python
# Create output directory
output_dir = Path('../outputs/notebook_results')
output_dir.mkdir(parents=True, exist_ok=True)

# Save point cloud
save_pointcloud(
    pc['points'],
    output_dir / 'pointcloud.ply',
    colors=pc['colors']
)

print(f"Results saved to {output_dir}")
```
